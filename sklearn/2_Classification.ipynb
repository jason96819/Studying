```python
import numpy as np
import pandas as pd
```

쉽게 이해하기 위해 iris 데이터 중 2가지로만 진행

```python
from sklearn.datasets import load_iris

data = load_iris()
df = pd.DataFrame(data=data.data, columns=data.feature_names)
df = df.iloc[:, :2]
df['target'] = data.target
df = df[df['target'] != 2]
df
```

```python
import matplotlib.pyplot as plt

df_class_0 = df[df['target'] == 0]
df_class_1 = df[df['target'] == 1]

plt.scatter(df_class_0.iloc[:, 0], df_class_0.iloc[:, 1], color='blue', label='target 0')
plt.scatter(df_class_1.iloc[:, 0], df_class_1.iloc[:, 1], color='red', label='target 1')

plt.grid()
plt.legend()
plt.show()
```
