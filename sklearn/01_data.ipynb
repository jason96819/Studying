# 정규화

```python
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

print('꽃받침 길이 평균 :', np.mean(df.iloc[:, 0]))
print('꽃받침 길이 분산 :', np.var(df.iloc[:, 0]))
print('='*20)
print('꽃받침 너비 평균 :', np.mean(df.iloc[:, 1]))
print('꽃받침 너비 분산 :', np.var(df.iloc[:, 1]))
print('='*20)
print('꽃잎 길이 평균 :', np.mean(df.iloc[:, 2]))
print('꽃잎 길이 분산 :', np.var(df.iloc[:, 2]))
print('='*20)
print('꽃잎 너비 평균 :', np.mean(df.iloc[:, 3]))
print('꽃잎 너비 분산 :', np.var(df.iloc[:, 3]))
```

```python
df.mean()
```

```python
df.var()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df)
iris_scaled = scaler.transform(df)

df_scaled = pd.DataFrame(data=iris_scaled, columns=iris.feature_names)

print('평균 :',df_scaled.mean())
print('분산 :',df_scaled.var())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(df)
iris_scaled_mm = scaler.transform(df)

df_mm_scaled = pd.DataFrame(data=iris_scaled_mm, columns=iris.feature_names)

print('최댓값 :',df_mm_scaled.min())
print('최솟값 :',df_mm_scaled.max())
```
