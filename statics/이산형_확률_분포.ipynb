## 베르누이 분포

베르누이의 확률 질량 함수는 다음과 같습니다.

$$P(X=x) = p^x \times (1-p)^{(1-x)},       (x = 0,1)​ $$

x는 0 또는 1일 때의 분포 입니다.

```python
# 성공 확률
p = 0.5

# 시행 횟수
n = 1

# 확률 질량 함수(PMF) 계산
pmf = [p ** x * (1 - p) ** (1 - x) for x in range(n + 1)]

# 결과 출력
for x in range(n + 1):
    print(f'X = {x}\t: P(X = {x}) = {pmf[x]}')
```

```python
plt.bar(range(n + 1), pmf)
plt.title('Bernoulli Distribution')
plt.xlabel('X')
plt.ylabel('Probability')
plt.show()
```

```python
plt.bar(range(n + 1), pmf)
plt.title('Bernoulli Distribution')
plt.xlabel('X')
plt.ylabel('Probability')
plt.show()
```

```python
from scipy.stats import bernoulli
```

```python
# 성공 확률
p = 0.3

# 시행 횟수
n = 1

# 베르누이 메서드
rv = bernoulli(p)

# 확률 질량 함수(PMF) 계산
x = range(n+1)
pmf = rv.pmf(x)

# 결과 출력
for i in range(n+1):
    print(f'X = {i}\t: P(X = {i}) = {pmf[i]}')
```

## 이항 분포

이항 분포의 확률 질량 함수는 다음과 같습니다.

$$P(X=x) = {n \choose x} p^x \times (1-p)^{(n-x)},       (x = 0,1)​ $$

x는 0 또는 1이고 n번 시행 했을때 의 분포 입니다.

```python
import math
```

```python
# 성공 확률
p = 0.5

# 시행 횟수
n = 10

# 확률 질량 함수(PMF) 계산
pmf = [(math.comb(n, x)) * (p ** x) * ((1 - p) ** (n - x)) for x in range(n+1)]

# 결과 출력
for x in range(n+1):
    print(f'X = {x}\t: P(X = {x}) = {pmf[x]}')
```

```python
plt.bar(range(n+1), pmf)
plt.title('Binomial Distribution')
plt.xlabel('X')
plt.ylabel('Probability')
plt.show()
```

SciPy의 binom 모듈을 사용하여 이항 분포의 확률 분포를 수행할 수 있습니다.

```python
from scipy.stats import binom
```

```python
def bin(n, p):
    # 이항 분포 메서드
    rv = binom(n, p)

    # 확률 질량 함수(PMF) 계산
    x = range(n+1)
    pmf = rv.pmf(x)

    # 결과 출력
    #for i in range(n+1):
    #    print(f'X = {i}\t: P(X = {i}) = {pmf[i]}')

    plt.bar(x, pmf)
    plt.title('Binomial Distribution')
    plt.xlabel('X')
    plt.ylabel('Probability')
    plt.show()
```

```python
bin(20, 0.3)
```

```python
bin(20, 0.9)
```

이항 분포에서 시행 횟수 n을 1로 하면 베르누이 분포가 됩니다.

```python
bin(1, 0.3)
```

반대로 이항 분포의 시행횟수를 무한대로 늘리면 연속형 정규 분포 형태를 보입니다.

## 기하 분포

기하 분포의 확률 질량 함수는 다음과 같습니다.

$$P(X=x) = (1-p)^{(x-1)} \times p$$

베르누이 시행을 계속 실패하다가 성공이 나오기까지 시행한 총횟수를 확률변수로 하는 분포


```python
# 성공 확률
p = 0.5

# 시행 횟수 즉, n번 만에 성공이 나왔다.
n = 10

# 확률 질량 함수(PMF) 계산
pmf = [p * (1 - p) ** (x - 1) for x in range(n+1)]

# 결과 출력
for x in range(n + 1):
    print(f'X = {x}\t: P(X = {x}) = {pmf[x]}')
```

```python
plt.bar(range(n + 1), pmf)
plt.title('Geometric Distribution')
plt.xlabel('X')
plt.ylabel('Probability')
plt.show()
```

SciPy의 geom 모듈을 사용하여 기하 분포의 확률 분포를 수행할 수 있습니다.

```python
from scipy.stats import geom
```

```python
def geo(n, p):
    # 시행 횟수 범위
    num = np.arange(1, n+1)

    # 기하 분포 생성
    rv = geom(p)

    # 확률 질량 함수(PMF) 계산
    pmf = rv.pmf(num)

    # 그래프 그리기
    plt.plot(num, pmf, marker='o', linestyle='-')

    # 그래프 타이틀과 축 레이블 설정
    plt.title('Geometric Distribution')
    plt.xlabel('X')
    plt.ylabel('Probability')

    # 그래프 보여주기
    plt.show()
```

```python
geo(10, 0.5)
```

p값에 따라 그래프가 어떻게 다르게 출력되는지 함수를 수정해 보겠습니다.

```python
def geo(n, p1, p2, p3, p4):
    # 시행 횟수 범위
    num = np.arange(1, n+1)

    p = [p1, p2, p3, p4]

    for i in range(1, 5):
        # 기하 분포 생성
        rv = geom(p[i-1])

        # 확률 질량 함수(PMF) 계산
        pmf = rv.pmf(num)

        # 그래프 그리기
        plt.plot(num, pmf, marker='o', linestyle='-', label=f'p ={p[i-1]}')

    # 그래프 타이틀과 축 레이블 설정
    plt.title('Geometric Distribution')
    plt.xlabel('X')
    plt.ylabel('Probability')
    plt.legend()

    # 그래프 보여주기
    plt.show()
```

```python
geo(10, 0.2, 0.4, 0.6, 0.8)
```

## 음이항 분포

음이항 분포의 확률 질량 함수는 다음과 같습니다.

$$P(X=x) = _{x+r-1}\rm C_x \times p^x \times (1-p)^r$$

r 번의 실패가 나오기까지 발생한 성공이 k 번일 확률




```python
def nbin_nm(r, p, n):
    # 성공 횟수 범위
    k = np.arange(n+1)

    # 확률 질량 함수 계산
    pmf = [(math.comb(x + r - 1, x)) * (p ** x) * ((1 - p) ** (r)) for x in k]

    # 결과 출력
    #for i in range(len(k)):
    #    print(f'K = {k[i]}\t: P(K = {k[i]}) = {pmf[i]}')

    # 그래프 그리기
    plt.plot(k, pmf, marker='o', linestyle='-')

    # 그래프 타이틀과 축 레이블 설정
    plt.title('Negative Binomial Distribution')
    plt.xlabel('K (Number of Successes)')
    plt.ylabel('Probability')

    # 그래프 보여주기
    plt.show()
```

```python
nbin_nm(10, 0.5, 50)
```

```python
nbin_nm(20, 0.5, 50)
```

SciPy의 nbinom 모듈을 사용하여 음이항 분포의 확률 분포를 수행할 수 있습니다.

```python
from scipy.stats import nbinom
```

```python
def nbin(r, p, n):
    # 성공 횟수 범위
    k = np.arange(n+1)

    # 음이항 분포 생성
    rv = nbinom(r, p)

    # 확률 질량 함수(PMF) 계산
    pmf = rv.pmf(k)

    # 결과 출력
    #for i in range(len(k)):
    #    print(f'K = {k[i]}\t: P(K = {k[i]}) = {pmf[i]}')

    # 그래프 그리기
    plt.plot(k, pmf, marker='o', linestyle='-')

    # 그래프 타이틀과 축 레이블 설정
    plt.title('Negative Binomial Distribution')
    plt.xlabel('K (Number of Successes)')
    plt.ylabel('Probability')

    # 그래프 보여주기
    plt.show()
```

```python
nbin(5, 0.5, 50)
```

시행 횟수와 성공 확률은 고정한뒤 실패횟수의 변화에 따른 그래프의 변화를 보겠습니다.

```python
def nbin_r(p=0.5, n=50):
    # 성공 횟수 범위
    k = np.arange(n+1)

    for r in range(1, 31, 6):
        # 음이항 분포 생성
        rv = nbinom(r, p)

        # 확률 질량 함수(PMF) 계산
        pmf = rv.pmf(k)

        # 그래프 그리기
        plt.plot(k, pmf, marker='o', linestyle='-', label=f'r = {r}')

    # 그래프 타이틀과 축 레이블 설정
    plt.title('Negative Binomial Distribution (p=0.5)')
    plt.xlabel('K (Number of Successes)')
    plt.ylabel('Probability')
    plt.legend()

    # 그래프 보여주기
    plt.show()
```

```python
nbin_r()
```

시행 횟수와 실패 횟수는 고정한뒤 성공 확률의 변화에 따른 그래프의 변화를 보겠습니다.

```python
def nbin_p(r=10, n=50):
    # 성공 횟수 범위
    k = np.arange(n+1)

    for p in np.arange(0.1, 0.8, 0.2):
        # 음이항 분포 생성
        rv = nbinom(r, p)

        # 확률 질량 함수(PMF) 계산
        pmf = rv.pmf(k)

        # 그래프 그리기
        plt.plot(k, pmf, marker='o', linestyle='-', label=f'p = {p:.1f}')

    # 그래프 타이틀과 축 레이블 설정
    plt.title('Negative Binomial Distribution (r=10)')
    plt.xlabel('K (Number of Successes)')
    plt.ylabel('Probability')
    plt.legend()

    # 그래프 보여주기
    plt.show()
```

```python
nbin_p()
```

## 푸아송 분포

음이항 분포의 확률 질량 함수는 다음과 같습니다.

$$P(X=x) = {\lambda^x \times e^{-\lambda} \over x!}$$

$$ \lambda = np$$

푸아송 분포는 이항분포에서 시행 횟수가 무한히 커지고, 성공 확률이 무한히 작아지는 경우에 유도되는 분포





```python
# 람다
lam = 3

# 시행 횟수 분포
n = np.arange(51)

# 푸아송 분포의 확률 질량 함수
pmf = [((lam ** x) * (np.exp(-lam))) / math.factorial(x) for x in n]

# 그래프 그리기
plt.plot(n, pmf, marker='o', linestyle='-')
plt.title('Poisson Distribution (lambda=3)')
plt.xlabel('x')
plt.ylabel('Probability')
plt.show()
```

SciPy의 poisson 모듈을 사용하여 푸아송 분포의 확률 분포를 수행할 수 있습니다.

```python
from scipy.stats import poisson
```

```python
def poi(lam, n):
    # 시행 횟수 분포
    num = np.arange(n+1)

    # 푸아송 분포의 확률 질량 함수 계산
    pmf = poisson.pmf(num, lam)

    # 그래프 그리기
    plt.plot(num, pmf, marker='o', linestyle='-', label=f'λ = {lam}')
    plt.xlabel('x')
    plt.ylabel('p')
    plt.title('Poisson Distribution')
    plt.legend()
    plt.show()
```

```python
poi(5, 50)
```

람다에 따른 푸아송 그래프의 변화량을 살펴보겠습니다.

```python
def poi_lam(n=100):
    # 람다 여러개
    lam = [5, 10, 20, 30, 40, 50, 60, 70]

    # 시행 횟수 분포
    num = np.arange(n+1)

    for i in lam:
        # 푸아송 분포의 확률 질량 함수 계산
        pmf = poisson.pmf(num, i)

        # 그래프 그리기
        plt.plot(num, pmf, marker='o', linestyle='-', label=f'λ = {i}')
    plt.xlabel('x')
    plt.ylabel('p')
    plt.title('Poisson Distribution')
    plt.legend()
    plt.show()
```

```python
poi_lam()
```

## 초기하 분포

음이항 분포의 확률 질량 함수는 다음과 같습니다.

$$P(X=x) = {_k\rm C_x \times _{m-k}\rm C_{n-x} \over _m\rm C_n}$$

크기가 m인 모집단에서 필요한 원소가 k 개가 있습니다. 이제 모집단에서 크기가 n인 표본을 뽑아 그 표본 안에 필요한 원소가 x 개 들어있을 확률분포





```python
from scipy.stats import hypergeom
```

```python
# 모집단 크기
m = 100

# 표본 집단 크기
n = 20

# 성공 원소 수
k = 5

# 특정 시행 횟수
x = np.arange(101)

# 확률 질량 함수 계산
pmf = hypergeom.pmf(k, m, n, x)

plt.plot(x, pmf, linestyle='-')
```

```python
def hyge(m=500, n=300, x=300):
    k_values = [10, 50, 100, 150, 200, 250, 300, 350, 400, 450]

    num = np.arange(x + 1)

    for k in k_values:
        pmf = hypergeom.pmf(num, m, n, k)

        plt.plot(num, pmf, linestyle='-', label=f'k = {k}')

    plt.xlabel('x')
    plt.ylabel('p')
    plt.title('Hypergeometric Distribution')
    plt.legend()
    plt.show()
```

```python
hyge()
```
