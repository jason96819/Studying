# 큰 수의 법칙

```python
# 필요한 라이브러리 import

import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# 평균이 100, 표준편차가 10인 백만개의 모집단 생성
population = np.random.normal(100, 10, 1000000)

# 정수값만 가져옴
#population = np.round(population)
```

```python
# 히스토그램 생성
hist_values, bin_edges = np.histogram(population, bins=np.arange(min(population), max(population) + 1))

# 히스토그램 시각화
plt.bar(bin_edges[:-1], hist_values, width=1, color='blue', alpha=0.7)
plt.title('Population Histogram')
plt.xlabel('Value')
plt.ylabel('Count')
plt.show()
```

```python
# 모집단에서 10개의 표본을 복원추출한다.
sample1 = np.random.choice(population, size=10, replace=True)
sample1
```

```python
# 표본평균
print('10개의 표본평균 : ',np.mean(sample1))
```

```python
# 모평균 - 표본평균 (평균의 차이)
print('10개의 표본의 평균과 모집단의 평균의 차이 : ', abs(100-np.mean(sample1)))
```

```python
# 모집단에서 20개의 표본을 복원추출하고 평균을 구한다.
sample2 = np.random.choice(population, size=20, replace=True)
print('20개의 표본평균 : ',np.mean(sample2))
print('20개의 표본의 평균과 모집단의 평균의 차이 : ',abs(100-np.mean(sample2)))
```

```python
# 모집단에서 30개의 표본을 복원추출하고 평균을 구한다.
sample3 = np.random.choice(population, size=30, replace=True)
print('30개의 표본평균 : ',np.mean(sample3))
print('30개의 표본의 평균과 모집단의 평균의 차이 : ',abs(100-np.mean(sample3)))
```

```python
diff_list = []

for i in range(10, 1001, 10):
    sample4 = np.random.choice(population, size=i, replace=True)
    #print(f'{i}개의 표본의 평균 : ',np.mean(sample4))
    #print(f'{i}개의 표본크기일 때 평균의 차이 : ', 100-np.mean(sample4))
    diff = abs(100 - np.mean(sample4))
    diff_list.append(diff)

sample_sizes = list(range(10, 1001, 10))

plt.figure(figsize=(12, 6))
plt.plot(sample_sizes, diff_list, linestyle='-')
plt.title('The difference between sample mean and population mean as a function of sample size.')
plt.xlabel('Sample Size')
plt.ylabel('DIFF(PM-SM)')
plt.grid(True)

plt.axhline(y=0, color='red', linestyle='--')

plt.show()
```

# 표본평균의 평균

1. 불확실성 감소: 표본을 하나만 사용하면 표본평균이 모평균에 얼마나 가까운지 정확히 알 수 없습니다. 여러 번 표본을 추출하고 평균을 계산하면 표본평균의 분포를 살펴봄으로써 추정치에 대한 불확실성을 감소시킬 수 있습니다.

2. 표본의 변동성 완화: 표본은 모집단을 완벽하게 대표하지 못할 수 있으며, 특정 표본이 우연히 극단적인 값으로 인해 편향될 수 있습니다. 여러 번 표본을 추출하고 평균을 계산하면 이러한 편향을 완화할 수 있습니다.

3. 추정치의 안정성: 여러 번 표본을 추출하고 평균을 내면 추정치가 더 안정적이 됩니다. 이는 특정 표본에 의존하지 않고 모집단을 대표하는 더 객관적인 추정치를 제공합니다.

4. 중심극한정리와 큰수의 법칙 활용: 중심극한정리와 큰수의 법칙에 따라 표본 평균의 분포는 표본의 크기가 충분히 크면 정규분포에 가까워지므로, 이를 통해 표본평균의 특성을 더 잘 이해하고 활용할 수 있습니다.

이를 통해 불확실성을 줄이고 추정치의 신뢰성을 높일 수 있습니다.

```python
# 통계학적으로 30개 이상의 표본은 큰표본이기에 30개의 표본 복원추출.
sample9 = np.random.choice(population, size=30)
```

```python
sample_mean_list = []

for _ in range(10):                                                              # 0~9까지 10번 반복한다.
    sample10 = np.random.choice(population, size=30)                               # 30개의 표본 복원추출.
    sample_mean = np.mean(sample10)                                                # 표본 30개의 평균 산출
    sample_mean_list.append(np.mean(sample10))                                     # sample_mean_list에 표본평균 삽입
```

```python
# sample_mean_list에는 총 10개의 표본평균의 평균값이 들어가있다.
sample_mean_list
```

```python
plt.hist(sample_mean_list, bins=30, color='blue', alpha=0.7)                     # 10개의 표본평균의 평균값 시각화
plt.title('Distribution of the sample mean')
plt.xlabel('Value')
plt.ylabel('Count')
plt.show()
```
