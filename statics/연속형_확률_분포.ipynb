# 연속형 확률 분포

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

## 연속형 확률 분포

## 균등 분포

```python
from scipy.stats import uniform
```

```python
a = 1  # 구간의 시작
b = 5  # 구간의 끝

# 그래프 범위 설정
x = np.linspace(a - 1, b + 1, 1000)

# 균등분포의 확률밀도함수 계산
pdf = uniform.pdf(x, loc=a, scale=b-a)

# 그래프 그리기
plt.plot(x, pdf)
plt.title('Uniform Distribution PDF')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

```python
# 균등분포 구간 설정
a = 1
b = 5

# 평균 및 분산 계산
mean = (a + b) / 2
variance = ((b - a)**2) / 12

# 확률밀도함수 계산
x = np.linspace(a - 1, b + 1, 1000)
pdf = uniform.pdf(x, loc=a, scale=b-a)

# 그래프 그리기
plt.plot(x, pdf, label='Uniform Distribution')
plt.axvline(x=mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
plt.title('Uniform Distribution PDF with Mean')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

print(f'Mean: {mean:.2f}')
print(f'Variance: {variance:.2f}')
```

## 정규 분포

```python
from scipy.stats import norm
```

```python
# 평균과 표준편차
mu = 0
sigma = 1

# 확률밀도함수 계산
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
pdf = norm.pdf(x, mu, sigma)

# 그래프 그리기
plt.plot(x, pdf)
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

```python
# 평균과 표준편차
mu = [0, 1, 2, 3, 4, 5]
sigma = 1

for i in mu:
    # 확률밀도함수 계산
    x = np.linspace(i - 3 * sigma, i + 3 * sigma, 100)
    pdf = norm.pdf(x, i, sigma)

    # 그래프 그리기
    plt.plot(x, pdf, label=f'mu = {i}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 평균과 표준편차
mu = 0
sigma = [1, 2, 3, 4, 5]

for i in sigma:
    # 확률밀도함수 계산
    x = np.linspace(mu - 3 * i, mu + 3 * i, 100)
    pdf = norm.pdf(x, mu, i)

    # 그래프 그리기
    plt.plot(x, pdf, label=f'sigma = {i}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.grid(True)
plt.legend()
plt.show()
```

## 지수 분포

```python
from scipy.stats import expon
```

```python
# 파라미터 lambda
lmbda = [0.1, 0.3, 0.5, 0.7, 0.9]

for i in lmbda:
    # 확률밀도함수 계산
    x = np.linspace(0, 10, 100)
    pdf = expon.pdf(x, scale=1/i)

    # 그래프 그리기
    plt.plot(x, pdf, label=f'lambda = {i}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution')
plt.grid()
plt.legend()
plt.show()
```

## 감마 분포

```python
from scipy.stats import gamma
```

```python
# 감마분포의 모수 설정
a = 2                   # 형태 모수
b = 1                   # 척도 모수

# 감마분포의 확률밀도함수 계산
x = np.linspace(0, 10, 1000)
pdf = gamma.pdf(x, a=a, scale=b)

# 그래프 그리기
plt.plot(x, pdf, label=f'Gamma Distribution (alpha={a}, beta={b})')
plt.title('Gamma Distribution PDF')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 감마분포의 모수 설정
a = [1, 2, 4, 8, 16]    # 형태 모수
b = 1                   # 척도 모수

# 감마분포의 확률밀도함수 계산
x = np.linspace(0, 30, 1000)

for i in a:
    pdf = gamma.pdf(x, a=i, scale=b)
    plt.plot(x, pdf, label=f'alpha={i}, beta={b}')

# 그래프 그리기
plt.title('Gamma Distribution PDF')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 감마분포의 모수 설정
a = 1                   # 형태 모수
b = [1, 2, 4, 8, 16]    # 척도 모수

# 감마분포의 확률밀도함수 계산
x = np.linspace(0, 30, 1000)

for i in b:
    pdf = gamma.pdf(x, a=a, scale=i)
    plt.plot(x, pdf, label=f'alpha={a}, beta={i}')

# 그래프 그리기
plt.title('Gamma Distribution PDF')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 감마분포의 모수 설정
a = [1, 2, 4, 8, 16]    # 형태 모수
b = [1, 2, 4, 8, 16]    # 척도 모수

# 감마분포의 확률밀도함수 계산
x = np.linspace(0, 30, 1000)

for i in a:
    for j in b:
        pdf = gamma.pdf(x, a=i, scale=j)
        plt.plot(x, pdf, label=f'alpha={i}, beta={j}')

# 그래프 그리기
plt.title('Gamma Distribution PDF')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

## 카이제곱 분포

```python
from scipy.stats import chi2
```

```python
# 자유도
kv = [2, 3, 5, 7, 9, 11]

# x 범위
x = np.linspace(0, 30, 500)

for k in kv:
    # 카이제곱 분포 확률밀도함수 계산
    pdf = chi2.pdf(x, k)

    # 그래프 그리기
    plt.plot(x, pdf, label=f'k = {k}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Chi-Square Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 자유도
k = 20

# x 범위
x = np.linspace(0, 60, 500)

# 카이제곱 분포 확률밀도함수 계산
pdf = chi2.pdf(x, k)

# 그래프 그리기
plt.plot(x, pdf, label=f'k = {k}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Chi-Square Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

## t 분포

```python
from scipy.stats import t
```

```python
# 자유도
k = [1, 3, 5, 10, 30]

# x 범위
x = np.linspace(-5, 5, 500)

for i in k:
    # t 분포 확률밀도함수 계산
    pdf = t.pdf(x, i)

    # 그래프 그리기
    plt.plot(x, pdf, label=f'k = {i}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('t Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

## f 분포

```python
from scipy.stats import f
```

```python
# 자유도
k1 = 5  # 분자 자유도
k2 = [1, 2, 4, 6, 8, 10]  # 분모 자유도

for i in k2:
    # x 범위
    x = np.linspace(0.01, 5, 500)

    # f 분포 확률밀도함수 계산
    pdf = f.pdf(x, k1, i)

    # 그래프 그리기
    plt.plot(x, pdf, label=f'k1 = {k1}, k2 = {i}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('F Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 자유도
k1 = [1, 2, 4, 6, 8, 10]  # 분자 자유도
k2 = 5  # 분모 자유도

for i in k1:
    # x 범위
    x = np.linspace(0.01, 5, 500)

    # f 분포 확률밀도함수 계산
    pdf = f.pdf(x, i, k2)

    # 그래프 그리기
    plt.plot(x, pdf, label=f'k1 = {i}, k2 = {k2}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('F Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
