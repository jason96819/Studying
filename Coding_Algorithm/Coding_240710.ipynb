## 문제 풀이

그래프 탐색 문제

BFS로 풀이.

그래프를 순회하다가 1인칸을 발견하면 BFS를 수행한다.

BFS를 수행하면, 해당지점을 0으로 변경하여 방문처리하고 상하좌우 또한 확인하여 방문처리

```python
from collections import deque
```

```python
number = int(input())
```

```python
dx = [-1, 1, 0, 0]      # 좌우
dy = [0, 0, -1, 1]      # 상하
```

```python
def bfs(a, b):
    q = deque()
    q.append((a, b))
    while q:
        x, y = q.popleft()

        for i in range(4):                                  # 상하좌우 이동
            nx = x + dx[i]
            ny = y + dy[i]

            if nx < 0 or nx >= M or ny < 0 or ny >= N:      # 좌표확인
                continue
            if array[nx][ny] == 1:                          # 배추가 심어져 있는 조건문
                q.append((nx, ny))                          # 방문 처리
                array[nx][ny] = 0                           # 0으로 변경
```

```python
for i in range(number):                                      # n번 반복
    M, N, K = map(int, input().split())
    array = [[0] * N for _ in range(M)]

    for j in range(K):
        x, y = map(int, input().split())
        array[x][y] = 1

    result = 0
    for a in range(M):
        for b in range(N):
            if array[a][b] == 1:                            # 배추가 심어져 있으면 bfs 수행
                bfs(a, b)
                result += 1
    print(result)
```

```python
# 최종답

from collections import deque

number = int(input())

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def bfs(a, b):
    q = deque()
    q.append((a, b))
    while q:
        x, y = q.popleft()

        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]

            if nx < 0 or nx >= M or ny < 0 or ny >= N:
                continue
            if array[nx][ny] == 1:
                q.append((nx, ny))
                array[nx][ny] = 0

for i in range(number):
    M, N, K = map(int, input().split())
    array = [[0] * N for _ in range(M)]

    for j in range(K):
        x, y = map(int, input().split())
        array[x][y] = 1

    result = 0
    for a in range(M):
        for b in range(N):
            if array[a][b] == 1:
                bfs(a, b)
                result += 1
    print(result)
```
